Filter rows in excel file with the DM key value


```python
import pandas as pd

def filter_excel_file() : 
    data = pd.read_excel('./Radiology_manual_annotations.xlsx')
    for row_index, row in data.iterrows() : 
        if row["Type"] == "DM" :
            data = data.drop(index= row_index) 

    data.to_excel("./Radiology_manual_anootations_filterd.xlsx" , index = False)
    
filter_excel_file()
```

convert image name

```python
def convert_image_name(image_name) :
    return image_name + ".jpg" 
```

Resize images


```python
from PIL import Image


def resize_image(input_image_path, output_image_path, new_size):
    print(input_image_path)
    image = Image.open(input_image_path)
    resized_image = image.resize(new_size)
    resized_image.save(output_image_path)



raw_input_path = r"./Subtracted_images_of_CDD_CESM/"
raw_output_path = r"./Resized_image/"
data = pd.read_excel('./Radiology_manual_anootations_filterd.xlsx')
new_size = (512, 512)  

for index,row in data.iterrows() : 
    input_path = raw_input_path + convert_image_name(row["Image_name"].rstrip())
    output_path = raw_output_path + convert_image_name(row["Image_name"])
    resize_image(input_image_path=input_path,output_image_path=output_path,new_size=new_size)

print("images successfully resized to 512px , 512px ")
```
